```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
    np.random.seed(seed)
    
    sample_ages = np.random.normal(loc=mu, scale=sigma, size=num_samples)
    sample_ages = np.round(sample_ages, decimals=0)
    
    return sample_ages
```

```python
sample = create_ages()
```

```python
sample
```

```python
plt.style.use('ggplot')
sns.histplot(sample, bins=20)
plt.show()
```

```python
sns.boxplot(data=sample)
plt.yticks([])
plt.show()
```

```python
df = pd.read_csv('Ames_Housing_Data.csv')
```

```python
df.head()
```

```python
mask = np.zeros_like(df.corr())
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(20, 15))
sns.heatmap(data=df.corr(), mask=mask, annot=True)
plt.show()
```

```python
df.corr()['SalePrice'].sort_values()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Overall Qual', y='SalePrice', data=df)
plt.title('Overall Qual Vs SalePrice', color='blue', size=16)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(x=df['Gr Liv Area'], y=df['SalePrice'], color='darkred')
plt.title('Gr Liv Area Vs SalePrice', size=16, color='blue')
plt.xlabel('Gr Liv Area', color='blue', size=14)
plt.ylabel('SalePrice',color='blue', size=14)
plt.show()
```
